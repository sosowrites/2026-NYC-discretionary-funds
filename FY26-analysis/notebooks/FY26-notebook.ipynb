```python
#importing pandas to work in pandas to clean and analyze this data!
import pandas as pd
```

```python
#in this notebook, I comb through the discretionary funding set aside by the City Council for FY26.
#Discretionary funding is the money set aside for *local community needs* and programs. 
#When combing through the data, I noticed that nearly $200,000 was set aside for programs that specifically focus on instilling Zionist values in New Yorkers, especially young New Yorkers.
#It's a tiny amount given the total FY26 budget, but I think worth asking what allegience to Israel has to do with local community needs. 
```

```python
#loading the data set
FY26_data = pd.read_csv('../data/NEW-Funded-Disclosure-FY2026.csv', encoding="latin-1")
```

```python
#taking a look at the first rows of the data set, all the discretionary funding for FY26
FY26_data.head()
```

```python
#looking at all the columns in the FY26 discretionary funds
FY26_data.columns
```

```python
#I know that the city council sets aside discretionary funds for organizations with the purpose to instill Zionist values in New Yorkers. Creating a data set with just those orgs.

contains_israel = FY26_data[FY26_data["Legal Name"].str.contains("israel", case=False, na=False)]
```

```python
#seeing the full list of orgs that might possibly foster support for Israel.
contains_israel
```

```python
#Did some research, Agudath Israel of America Community Services is focused on supporting Jewish values more broadly and should be removed.
#This method creates a data set of every single row that does not have a string that contains "Agudath Israel" in the "Legal Name" column
contains_israel[~contains_israel["Legal Name"].str.contains("Agudath Israel")]
```

```python
#this method creates a new data set with all the rows that do not have "Agudath Israel of America Community Services, Inc." in the "Legal Name" column
israel_list1 = contains_israel[contains_israel["Legal Name"]!="Agudath Israel of America Community Services, Inc."]
```

```python
#checking my new data set
israel_list1
```

```python
#Hope of Israel Senior Citizens Center, Inc. should also be removed. Doing so with the same method I used earlier.
israel_list2 = israel_list1[~israel_list1["Legal Name"].str.contains("Hope of Israel Senior Citizens")]
```

```python
#checking my new list.
israel_list2
```

```python
#Rabbi Israel Meyer HaKohen Rabbinical Seminary should also be removed. Doing so with the same method I used earlier.
israel_list3 = israel_list2[~israel_list2["Legal Name"].str.contains("Rabbi Israel Meyer")]
```

```python
#Checking my final list
israel_list3
```
